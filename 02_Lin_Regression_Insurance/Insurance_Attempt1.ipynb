We know our data is right skewed hence we normalise it first.

**`But why do we normalise?`**

If the dependent variable is skewed, it can cause the regression model to assign too much weight to extreme values, leading to biased estimates of the coefficients and inaccurate predictions. This can be particularly problematic if the skewness is caused by outliers, as they can have a significant impact on the model's predictions.

There are several ways to normalize a skewed dependent variable, including:

**`Logarithmic transformation:`** One way to normalize a positively skewed variable is to take the logarithm of the values. This can help to reduce the impact of extreme values and make the distribution more symmetrical. To do this, you can use the numpy log function.

**`Box-Cox transformation:`** The Box-Cox transformation is a more general transformation that can be used to normalize a variable regardless of its skewness. It involves raising the variable to a power that is estimated from the data. The optimal value of the power can be estimated using the scipy.stats.boxcox function.

**`Z-score normalization:`** Z-score normalization involves subtracting the mean and dividing by the standard deviation of the variable, which transforms the variable into a standard normal distribution with a mean of 0 and a standard deviation of 1. This can be done using the scipy.stats.zscore function

```python
df['log_charges'] = pd.np.log(df.charges)
```

Lets move to gender column

```python
df.sex = df.sex.map({'female':0, 'male':1})
```

```python
df.smoker = df.smoker.map({'yes':1, 'no':0})
```

```python
df_region_encoded = pd.get_dummies(df['region'], prefix='region')
df = pd.concat([df, df_region_encoded], axis=1)
```

```python
df.drop(['region', 'charges'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

y = df[['log_charges']]
X = df[['age', 'sex', 'bmi', 'children', 'smoker',
       'region_northeast', 'region_northwest', 'region_southeast',
       'region_southwest']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
import statsmodels.api as sm

X_train = sm.add_constant(X_train)
y_train = y_train.values

model = sm.OLS(y_train, X_train)
results = model.fit()
print(results.summary())
```

Till now we were just doing modeling but do you really feel ML is this simple? Let us see the what challenges we face with ML.

**`Assumptions of Linear regression`**


Linear regression makes several assumptions about the data that are used to build the model. Here are the four main assumptions of linear regression and their potential solutions:

**`Linearity`**: Linear regression assumes that the relationship between the independent and dependent variables is linear. If this assumption is not met, the model may not fit the data well and may not be useful for prediction. One potential solution is to transform the data, such as by taking the logarithm or square root of the variables, to achieve a more linear relationship.

**`Independence`**: Linear regression assumes that the observations are independent of each other. This means that there should be no correlation between the residuals (i.e., the differences between the predicted and actual values) of the model. One potential solution is to ensure that the data is collected in a way that avoids dependencies, such as by using random sampling.

**`Homoscedasticity`**: Linear regression assumes that the variance of the residuals is constant across all values of the independent variables. If this assumption is not met, the model may not fit the data well and may not be useful for prediction. One potential solution is to transform the dependent variable, such as by taking the logarithm or square root of the variable, to achieve a more uniform variance.

**`Normality`**: Linear regression assumes that the residuals follow a normal distribution. If this assumption is not met, the model may not fit the data well and may not be useful for prediction. One potential solution is to transform the dependent variable, such as by taking the logarithm or square root of the variable, to achieve a more normal distribution. Additionally, non-parametric methods, such as random forests or support vector regression, can be used as an alternative to linear regression when normality assumptions are not met.

It is important to note that these assumptions are not always strictly necessary for linear regression to be useful, and some deviations from these assumptions can be tolerated depending on the specific problem and the size of the dataset. However, it is important to be aware of these assumptions and potential solutions in order to build accurate and reliable models.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["feature"] = X.columns
```

```python
model = sm.OLS(y_train, X_train_t)
results = model.fit()
print(results.summary())
```

```python
residuals = results.resid

# Create a normal probability plot of the residuals
fig = sm.qqplot(residuals, line='s')
plt.show()
```
